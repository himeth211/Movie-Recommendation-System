```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv('movies.csv')
# print(data.isnull().sum())

relevant = data[['title', 'genre', 'rating', 'votes', 'stars', 'description']]
columns = data[['genre', 'description', 'stars']]
# relevant = columns.drop(['description', 'genre'], axis=1)

cv = CountVectorizer(max_features=9957, stop_words='english')
vector = cv.fit_transform(columns.astype(str).agg(' '.join, axis=1)).toarray()
similarity = cosine_similarity(vector)
```

```python
movie_title = input("Enter Movie Title: ")
movie_title_to_find = movie_title.title()

if movie_title_to_find in relevant['title'].str.title().values:
    movie_index = relevant[relevant['title'].str.title() == movie_title_to_find].index[0]

    # Get the similarity scores for the selected movie
    movie_similarities = list(enumerate(similarity[movie_index]))

    # Sort movies by similarity score in descending order
    distance = sorted(movie_similarities, reverse=True, key=lambda x: x[1])

    print(f"Top 5 similar movies to '{movie_title_to_find}':")
    for i in range(2, min(7, len(distance))):
        similar_movie_idx = distance[i][0]
        similarity_score = distance[i][1]
        similar_movie_title = relevant.loc[similar_movie_idx, 'title']
        print(f"  {i-1}. {similar_movie_title} (Similarity: {similarity_score:.2f})")
else:
    print(f"Movie '{movie_title_to_find}' not found in the dataset.")
```
